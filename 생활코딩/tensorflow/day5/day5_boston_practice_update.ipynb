```python
import tensorflow as tf
import pandas as pd
```

```python
# 데이터 불러오기
filename = "boston.csv"
data = pd.read_csv(filename)
```

```python
# 데이터 분리
iv = data.iloc[:, :-1]
# dv = data.iloc[:, -1]
dv = data[['medv']]
```

```python
print(iv.shape, dv.shape)
```

```python
# 모델 구조 생성
X = tf.keras.layers.Input(shape=[13]) # iv 특성 수

##### 히든레이어 
# H = tf.keras.layers.Dens(8, activation='swish')(X) # 한번에 작성하는 방법
# 아래와 같이 하는 이유는 BatchNormailztion을 추가하기 위하여
# 추가이후 loss 는 12 -> 9
H = tf.keras.layers.Dense(8)(X)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)

H = tf.keras.layers.Dense(8)(H)
H = tf.keras.layers.BatchNormalization()(H)
H = tf.keras.layers.Activation('swish')(H)


Y = tf.keras.layers.Dense(1)(H) # dv 특성 수 
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
# verbose를 통해 러닝결과가 출력여부 결정
# 10000, 1000, 10 러닝 진행결과 23 아래로 잘 내려가지 않음
# 추가적인 러닝 방안 필요
model.fit(iv, dv, epochs=10000, verbose=0)
model.fit(iv, dv, epochs=10, verbose=1)
```

```python
# 모델 확인
model.summary()
```

```python
for x, y in zip(model.predict(iv[:5]), dv.medv[:5]):
    print("예측 -> {} , 결과 -> {}".format(round(x[0],1), y))
```
